```python
# 绘图全局设置
sns.set(
    style='darkgrid',
    font='Times New Roman',
    font_scale=2
)
ROW_NUM, COL_NUM = 2, 5
plt.figure(dpi=DPI)
fig, ax_arr = plt.subplots(ROW_NUM, COL_NUM, sharex='col', figsize=(20, 15))  # sharex、y设置子图共享x、y轴坐标
fig.subplots_adjust(hspace=0.2, wspace=0.32)  # 设置子图的横纵间距

for index, s in enumerate(df.iteritems()):
    cur_df = pd.concat([s[1], affix_info], axis=1)
    # 计算行列数
    row_num = index // COL_NUM
    col_num = index - row_num * COL_NUM
    cur_ax = ax_arr[row_num][col_num]
    cur_plot = sns.boxplot(
        x="River",
        y=s[0],
        data=cur_df,
        width=0.3,
        linewidth=3,
        dodge=True,
        palette="Set1",
        ax=cur_ax
    )
    # 子图图名、x轴、y轴、图例
    cur_ax.tick_params(axis='x', rotation=90)
    if len(df.columns) - (index + 1) >= COL_NUM:
        cur_plot.set_xlabel(None)
    cur_plot.set_title(f'${s[0]}$({unit[s[0]]})')
    cur_plot.set_ylabel(None)

# 保存图片
fig.savefig(EXPORT_PATH + '平水期不同入湖口环境因子变化图' + IMG_TYPE, dpi=DPI, bbox_inches='tight')
```

## 南淝河枯水期与丰水期对比

```python
# df[df.index.str.startswith('L-N') | df.index.str.startswith('W-N')]
df = pd.concat([overlying_water_env_df.copy(), sediment_env_df.iloc[:, :2]], axis=1)
df = df[df.index.str.startswith('D-N') | df.index.str.startswith('W-N')]
df = df[
    ['W-TN', 'W-NO_{3}^{-}', 'W-NO_{2}^{-}', 'W-NH_{4}^{+}', 'W-COD', 'W-DO', 'W-pH', 'W-T',
     'S-TN', 'S-TP']]
df
```

```python
# 绘图全局设置
sns.set(
    style='darkgrid',
    font='Times New Roman',
    font_scale=2
)
ROW_NUM, COL_NUM = 2, 5
plt.figure(dpi=DPI)
fig, ax_arr = plt.subplots(ROW_NUM, COL_NUM, sharex='col', figsize=(20, 15))  # sharex、y设置子图共享x、y轴坐标
fig.subplots_adjust(hspace=0.2, wspace=0.32)  # 设置子图的横纵间距

for index, s in enumerate(df.iteritems()):
    cur_df = pd.concat([s[1], affix_info], axis=1)
    cur_df.drop(index=cur_df[cur_df['Period'] == 'Level Season'].index, inplace=True)
    cur_df['Period'] = cur_df['Period'].cat.remove_unused_categories()

    # 计算行列数
    row_num = index // COL_NUM
    col_num = index - row_num * COL_NUM
    cur_ax = ax_arr[row_num][col_num]
    cur_plot = sns.boxplot(
        x="Period",
        y=s[0],
        data=cur_df,
        width=0.3,
        linewidth=3,
        dodge=True,
        palette="Set1",
        ax=cur_ax
    )
    # 子图图名、x轴、y轴、图例
    cur_ax.tick_params(axis='x', rotation=90)
    if len(df.columns) - (index + 1) >= COL_NUM:
        cur_plot.set_xlabel(None)
    cur_plot.set_title(f'${s[0]}$({unit[s[0]]})')
    cur_plot.set_ylabel(None)

# 保存图片
fig.savefig(EXPORT_PATH + '南淝河-枯水期&丰水期-环境因子变化图' + IMG_TYPE, dpi=DPI, bbox_inches='tight')
```
